# Linear Support Vector Classifier (Tuned)
No data preprocessing.

```python
import sys
sys.path.append("../../../")

import modules
import numpy as np
from sklearn import model_selection, svm, metrics

mydata = modules.NT("../../../data/b100d30.csv")
x_train, x_test, y_train, y_test = model_selection.train_test_split(mydata.data, mydata.target, test_size=0.4, random_state=508)
```

```python
svc = svm.LinearSVC()
param_grid = {"penalty": ['l1', 'l2'],
              "loss": ['hinge', 'squared_hinge'],
              "dual": [True, False],
              "tol": [1e-4, 1e-5, 1e-3, 1e-2, 1e-1],
              "C": [0.001, 0.01, 0.1, 0.25, 0.5, 1]}
search = model_selection.RandomizedSearchCV(estimator=svc, param_distributions=param_grid, n_jobs=-1, n_iter=20, verbose=1)
search.fit(x_train, y_train)
clf = search.best_estimator_
print(clf)
```

```python
y_pred = clf.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred))
print(y_test, y_pred)
```
